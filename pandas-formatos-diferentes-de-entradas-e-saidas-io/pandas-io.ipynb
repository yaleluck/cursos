# Criando os nomes

```python
import pandas as pd
```

```python
# importando um arquivo json
names_f = pd.read_json('https://guilhermeonrails.github.io/nomes_ibge/nomes-f.json')
names_f
```

```python
# importando um arquivo json
names_m = pd.read_json('https://guilhermeonrails.github.io/nomes_ibge/nomes-m.json')
names_m
```

```python
names_f.sample(5) # escolhendo randomicamente 5 nomes
```

```python
names_m.sample(5)
```

```python
# printando a soma das quantidades de nomes de cada lista
print('Quantidade de nomes: ' + str(len(names_f) + len(names_m)))
```

```python
frames = [names_f, names_m] # criando uma lista formada pelos 2 DataFrames
frames
```

```python
names = pd.concat(frames)['nome'].to_frame() # concatenando a variável 'nome' dos dois DataFrames em um novo data frame
names
```

```python
names.sample(5)
```

# Incluindo Id dos alunos

```python
import numpy as np
```

```python
np.random.seed(123)
```

```python
total_students = len(names) # criando uma variável com o nº total de alunos
total_students
```

```python
names['id_aluno'] = np.random.permutation(total_students) + 1
# adiciona ao DataFrame names a variável 'id_aluno' que possui um valor aleatório com valor máximo de 400
```

```python
names.sample(5)
```

```python
domains = ['@dominiodoemail.com.br', '@servicodoemail.com'] # criando a variável domains com os domínios de emails possíveis dos alunos
names['dominio'] = np.random.choice(domains, total_students) # escolhendo randomicamente entre os emails contidos em domains, um email para cada aluno do total de alunos (400)
```

```python
names
```

```python
names['email'] = names.nome.str.cat(names.dominio).str.lower() # criando a nova variável (coluna) concatenando cada nome com o seu respectivo dominio de email.
# .str converte em string, .cat concatena, .lower deixa os caraceres minúsculos (lowercase)
names
```

# Criando a tabela cursos 

```python
courses_list = ['Lógica de programação', 'Java para Web', 'C# para Web', 'Ruby on Rails', 'Cursos de Python', 'PHP com MySql', '.NET para web', 'Novas integrações com Java', 'TDD com Java', 'Código limpo com C#', 'Preparatório para certificação Java', 'Hardware básico', 'Persistência com .NET', 'Desenvolvendo jogos', 'Análise de dados', 'Estatística básica', 'Internet das coisas', 'Programação funcional', 'Boas práticas em Java', 'Orientação a objetos com Java']
courses_list
```

```python
courses = pd.DataFrame(courses_list)
courses.columns = ['Nome_do_Curso']
courses
```

# Alterando o index de cursos

```python
# outra forma de alterar o nome de uma variavel do DataFrame == courses = courses.rename(columns = {'0' : 'Nome_do_Curso})
```

```python
courses['id'] = courses.index + 1 # criando uma variável 'id' para o DataFrame courses
courses
```

```python
courses = courses.set_index('id') # atribuindo os valores da variável 'id' como índice do DataFrame
courses
```

# Matriculando os alunos nos cursos

```python
names['matriculas'] = np.ceil(np.random.exponential(size = total_students) * 1.5).astype(int) # criando a variável 'matricula' no DataFrame names utilizando valores randomicos exponenciais para o total de alunos (size)
# .ceil arredonda os valores para cima (>1)
```

```python
names
```

# Selecionando cursos

```python
total_registrations = []
x = np.random.rand(20)
prob = x / sum(x)
```

```python
for index, row in names.iterrows(): # .iterrows realiza a "iteração" linha por linha no DataFrame
    id = row.id_aluno
    regist = row.matriculas
    for i in range(regist):
        reg = (id, np.random.choice(courses.index, p = prob))
        total_registrations.append(reg)

registrations = pd.DataFrame(total_registrations, columns = ['id_aluno', 'id_curso'])
```

```python
registrations
```

```python
registrations.groupby('id_curso').count().join(courses['Nome_do_Curso']).rename(columns = {'id_aluno' : 'Quantidade_de_Alunos'}) # criando um agrupamento por 'id_curso' e realizando a contagem de quantos elementos foram agrupados por agrupamento com o .count()
```

```python
regist_per_course = registrations.groupby('id_curso').count().join(courses['Nome_do_Curso']).rename(columns = {'id_aluno' : 'Quantidade_de_Alunos'})
```

```python
regist_per_course
```
