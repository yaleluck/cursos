```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('./aluguel.csv', sep = ';')
dados
```

```python
area = plt.figure() # criando uma área para colocar os gráficos
```

```python
graph1 = area.add_subplot(2, 2, 1) #.add_subplot(2, 2, 1) estaremos adicionando um gráfico em uma área com 2 linhas, 2 colunas e esse gráfico estará na 1º posição
graph2 = area.add_subplot(2, 2, 2)
graph3 = area.add_subplot(2, 2, 3)
graph4 = area.add_subplot(2, 2, 4)
```

```python
graph1.scatter(dados.Valor, dados.Area) # .scatter é um gráfico de espalhamento de acordo com determinados parâmetros
graph1.set_title('Valor X Área')

graph2.hist(dados.Valor) # .hist cria um gráfico de histograma
graph2.set_title('Histograma')

dados_graph3 = dados.Valor.sample(100) # selecionando aleatoriamente 100 registros da variável Valor do dataset
dados_graph3.index = range(dados_graph3.shape[0]) # corrigindo os índices dos dados randômicos corrigidos
graph3.plot(dados_graph3) # .plot cria uma gráfico simples
graph3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor'] # criando um grupo com a variável Tipo de acordo com os valores de Valor
label = grupo.mean().index # os índices serão os Tipos de imóveis
valores = grupo.mean().values
graph4.bar(label, valores) # .bar cria um gráfico de barras
graph4.set_title('Valor Médio por Tipo')
```
