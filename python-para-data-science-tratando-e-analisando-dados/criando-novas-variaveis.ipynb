## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('./aluguel-residencial.csv', sep = ';')
```

```python
dados
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
# criando uma nova variável no DataFrame
```

```python
dados
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
# criando uma nova variável no DataFrame
```

```python
dados
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
# arredondando para 2 casas decimais
```

```python
dados
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Valor m2']).round(2)
```

```python
dados
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
# criando uma nova variável no DataFrame dados utilizando uma função lambda aplicada no método .apply da variável 'Tipo' no DataFrame dados.
```

```python
dados
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux
```

```python
del dados_aux['Valor Bruto'] # del deleta uma variável indicada do DataFrame
```

```python
dados_aux
```

```python
dados_aux.pop('Valor Bruto m2') # .pop retira uma variável indicada do DataFrame, exibindo essa retirada.
```

```python
dados_aux
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True) # .drop exclui variáveis do DataFrame podendo ser selecionado o eixo excluído (0 = linha, 1  = coluna)
```

```python
dados
```

```python
dados.to_csv('./aluguel-residencial.csv', sep = ';', index = False)
```
