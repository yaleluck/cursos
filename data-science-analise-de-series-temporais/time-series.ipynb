```python
def plotar(titulo, labelx, labely, x, y, dataset):
    sns.set_palette('Accent') # selecionando um conjunto de cores para o gráfico.
    sns.set_style('darkgrid') # adicionando o grid na área do gráfico.
    ax = sns.lineplot(x = x, y = y, data = dataset) # armazenando o gráfico em uma variável.
    ax.figure.set_size_inches(12, 6) # alterando as propriedades da figura do gráfico gerado com .figure e definindo a altura e largura do gráfico.
    ax.set_title(titulo, loc = 'left', fontsize = '18') # definindo o título do gráfico, sua localização e tamanho.
    ax.set_xlabel(labelx, fontsize = '14') # definindo o rótulo do eixo X e o seu tamanho.
    ax.set_ylabel(labely, fontsize = '14') # definindo o rótulo do eixo Y e o seu tamanho.
    ax = ax
```

```python
plotar('Aumento das Vendas da Alucar de 2017 a 2018', 'Tempo (Meses)', 'Aumento (Nº de vendas)', 'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff() # criando uma nova coluna no DataFrame com a aceleração do aumento do número de vendas de um mês em relação ao anterior [n - (n-1)]
alucar
```

```python
plotar('Aceleração das Vendas da Alucar de 2017 a 2018', 'Tempo', 'Aceleração', 'mes', 'aceleracao', alucar)
```

```python
plt.figure(figsize = (16, 12))
ax = plt.subplot(3, 1, 1)
ax.set_title('Análise de Vendas da Alucar de 2017 a 2018', fontsize = 18, loc = 'left')
sns.lineplot(x = 'mes', y = 'vendas', data = alucar)
plt.subplot(3, 1, 2)
sns.lineplot(x = 'mes', y = 'aumento', data = alucar)
plt.subplot(3, 1, 3)
sns.lineplot(x = 'mes', y = 'aceleracao', data = alucar)
ax = ax

# plotando vários gráficos em uma única figura, separando-os por áreas.
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    plt.figure(figsize = (16, 12))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 18, loc = 'left')
    sns.lineplot(x = x, y = y1, data = dataset)
    plt.subplot(3, 1, 2)
    sns.lineplot(x = x, y = y2, data = dataset)
    plt.subplot(3, 1, 3)
    sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
```

```python
plot_comparacao('mes', 'vendas', 'aumento', 'aceleracao', alucar, 'Análise de Vendas da Alucar 2017 a 2018')
```

```python
from pandas.plotting import autocorrelation_plot # importando o método autocorrelation_plot da biblioteca pandas.
```

```python
ax = plt.figure(figsize = (12, 6)) # criando a área para a figura
ax.suptitle('Correlação das vendas', fontsize = 18, x = 0.26, y = 0.95) # criando o subtítulo da figura e definindo a posição X e Y do titulo.
autocorrelation_plot(alucar['vendas']) # gerando o gráfico de correlação com os dados de 'vendas'
ax = ax
```

## Analisando a sazonalidade

```python
sales_per_day['dia_da_semana'] = sales_per_day['dia'].dt.day_name() # criando uma nova coluna em sales_per_day com o 'dia_da_semana' de cada 'dia'.
```

```python
sales_per_day
```

```python
sales_per_day['dia_da_semana'].unique() # retornando um array com todos os dados únicos em 'dia_da_semana'.
```

```python
translated_ptbr_days = {'Monday': 'Segunda', 'Tuesday': 'Terça', 'Wednesday': 'Quarta', 'Thursday': 'Quinta', 'Friday': 'Sexta', 'Saturday': 'Sábado', 'Sunday': 'Domingo'} # criando um dicionário com as traduções que cada dia da semana em inglês deverá receber.
```

```python
sales_per_day['dia_da_semana'] = sales_per_day['dia_da_semana'].map(translated_ptbr_days) # mapeando a coluna 'dia_da_semana' com .map() e substituindo seus valores através do dict translated_ptbr_days.
```

```python
sales_per_day
```

## Agrupando os dias

```python
stacked_sales = sales_per_day.groupby('dia_da_semana')['vendas', 'aumento', 'aceleração'].mean().round() # agrupando os dados do DataFrame sales_per_day pelo valor de cada 'dia_da_semana' e em seguida calculando a média dos valores agrupados e armazenando o resultado em stacked_sales
```

```python
stacked_sales
```

Observa-se que em feriados e finais de semana, as vendas caem.

OBS: Não devemos analisar a Time Series quando os dados não apresentam a variável Tempo ou quando as variáveis ou valores são constantes!

# Statsmodel

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
results = seasonal_decompose(chocolura['vendas'], period = 1, extrapolate_trend = 'freq') # como os valores em 'vendas' não são do tipo DateTime, precisa-se utilizar o period e para visualizar a frequência, usa-se extrapolate_trend = 'freq'.
ax = results.plot()
```

```python
results = seasonal_decompose(chocolura['vendas'], period = 2, extrapolate_trend = 'freq') # como os valores em 'vendas' não são do tipo DateTime, precisa-se utilizar o period e para visualizar a frequência, usa-se extrapolate_trend = 'freq'.
ax = results.plot()
```

```python
results = seasonal_decompose(chocolura['vendas'], period = 3, extrapolate_trend = 'freq') # como os valores em 'vendas' não são do tipo DateTime, precisa-se utilizar o period e para visualizar a frequência, usa-se extrapolate_trend = 'freq'.
ax = results.plot()
```

```python
observation = results.observed
trend = results.trend
seasonality = results.seasonal
noise = results.resid
```

```python
data = ({
    'observação': observation,
    'tendência': trend,
    'sazonalidade': seasonality,
    'ruído': noise
})

results = pd.DataFrame(data)
results
```

```python
plot_comparacao(results.index, 'observação', 'tendência', 'sazonalidade', results, 'Exemplo de Statsmodels')
```
