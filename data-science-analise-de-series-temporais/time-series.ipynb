```python
def plotar(titulo, labelx, labely, x, y, dataset):
    sns.set_palette('Accent') # selecionando um conjunto de cores para o gr√°fico.
    sns.set_style('darkgrid') # adicionando o grid na √°rea do gr√°fico.
    ax = sns.lineplot(x = x, y = y, data = dataset) # armazenando o gr√°fico em uma vari√°vel.
    ax.figure.set_size_inches(12, 6) # alterando as propriedades da figura do gr√°fico gerado com .figure e definindo a altura e largura do gr√°fico.
    ax.set_title(titulo, loc = 'left', fontsize = '18') # definindo o t√≠tulo do gr√°fico, sua localiza√ß√£o e tamanho.
    ax.set_xlabel(labelx, fontsize = '14') # definindo o r√≥tulo do eixo X e o seu tamanho.
    ax.set_ylabel(labely, fontsize = '14') # definindo o r√≥tulo do eixo Y e o seu tamanho.
    ax = ax
```

```python
plotar('Aumento das Vendas da Alucar de 2017 a 2018', 'Tempo (Meses)', 'Aumento (N¬∫ de vendas)', 'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff() # criando uma nova coluna no DataFrame com a acelera√ß√£o do aumento do n√∫mero de vendas de um m√™s em rela√ß√£o ao anterior [n - (n-1)]
alucar
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    plt.figure(figsize = (16, 12))
    ax = plt.subplot(3, 1, 1)
    ax.set_title(titulo, fontsize = 18, loc = 'left')
    sns.lineplot(x = x, y = y1, data = dataset)
    plt.subplot(3, 1, 2)
    sns.lineplot(x = x, y = y2, data = dataset)
    plt.subplot(3, 1, 3)
    sns.lineplot(x = x, y = y3, data = dataset)
    ax = ax
```

```python
plot_comparacao('mes', 'vendas', 'aumento', 'aceleracao', alucar, 'An√°lise de Vendas da Alucar 2017 a 2018')
```

## Analisando a sazonalidade

```python
sales_per_day['dia_da_semana'] = sales_per_day['dia'].dt.day_name() # criando uma nova coluna em sales_per_day com o 'dia_da_semana' de cada 'dia'.
```

```python
sales_per_day
```

```python
sales_per_day['dia_da_semana'].unique() # retornando um array com todos os dados √∫nicos em 'dia_da_semana'.
```

```python
translated_ptbr_days = {'Monday': 'Segunda', 'Tuesday': 'Ter√ßa', 'Wednesday': 'Quarta', 'Thursday': 'Quinta', 'Friday': 'Sexta', 'Saturday': 'S√°bado', 'Sunday': 'Domingo'} # criando um dicion√°rio com as tradu√ß√µes que cada dia da semana em ingl√™s dever√° receber.
```

```python
sales_per_day['dia_da_semana'] = sales_per_day['dia_da_semana'].map(translated_ptbr_days) # mapeando a coluna 'dia_da_semana' com .map() e substituindo seus valores atrav√©s do dict translated_ptbr_days.
```

```python
sales_per_day
```

## Agrupando os dias

```python
stacked_sales = sales_per_day.groupby('dia_da_semana')['vendas', 'aumento', 'acelera√ß√£o'].mean().round() # agrupando os dados do DataFrame sales_per_day pelo valor de cada 'dia_da_semana' e em seguida calculando a m√©dia dos valores agrupados e armazenando o resultado em stacked_sales
```

```python
stacked_sales
```

Observa-se que em feriados e finais de semana, as vendas caem.

# Statsmodel

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
results = seasonal_decompose(chocolura['vendas'], period = 1, extrapolate_trend = 'freq') # como os valores em 'vendas' n√£o s√£o do tipo DateTime, precisa-se utilizar o period e para visualizar a frequ√™ncia, usa-se extrapolate_trend = 'freq'.
ax = results.plot()
```

```python
results = seasonal_decompose(chocolura['vendas'], period = 2, extrapolate_trend = 'freq') # como os valores em 'vendas' n√£o s√£o do tipo DateTime, precisa-se utilizar o period e para visualizar a frequ√™ncia, usa-se extrapolate_trend = 'freq'.
ax = results.plot()
```

```python
results = seasonal_decompose(chocolura['vendas'], period = 3, extrapolate_trend = 'freq') # como os valores em 'vendas' n√£o s√£o do tipo DateTime, precisa-se utilizar o period e para visualizar a frequ√™ncia, usa-se extrapolate_trend = 'freq'.
ax = results.plot()
```

```python
observation = results.observed
trend = results.trend
seasonality = results.seasonal
noise = results.resid
```

```python
data = ({
    'observa√ß√£o': observation,
    'tend√™ncia': trend,
    'sazonalidade': seasonality,
    'ru√≠do': noise
})

results = pd.DataFrame(data)
results
```

```python
plot_comparacao(results.index, 'observa√ß√£o', 'tend√™ncia', 'sazonalidade', results, 'Exemplo de Statsmodels')
```
