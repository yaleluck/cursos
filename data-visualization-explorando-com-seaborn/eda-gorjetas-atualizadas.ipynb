## Teste de Hipótese

**H<sup>null</sup>**
>**A distribuição da taxa da gorjeta é a mesma nos dois grupos.**

**H<sup>alt</sup>**
>**A distribuição da taxa da gorjeta não é a mesma nos dois grupos.**

```python
from scipy.stats import ranksums
```

```python
dessert = tips.query("sobremesa == 'Sim'").porcentagem # criando uma variável para armazenar o retorno da query que verifica os dados onde a valor de 'sobremesa' é 'Sim'.
dessert
```

```python
no_dessert = tips.query("sobremesa == 'Não'").porcentagem # criando uma variável para armazenar o retorno da query que verifica os dados onde a valor de 'sobremesa' é 'Não'.
no_dessert
```

```python
ranksums(dessert, no_dessert)
```

Como pvalue=0.5266635660124415 é maior do que 0.05, descartamos a hipótese alternativa e mantemos a hipótese nula. Apesar de os dois grupos terem dado gorjetas de valores diferentes entre si, a diferença é insignificante do ponto de vista matemático. Ou seja, não poderemos dizer que a população geral será diferente também.

```python
r = ranksums(dessert, no_dessert)
print('O valor do p-value é {}'.format(r.pvalue.round(2)))
```
